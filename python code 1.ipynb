```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Читання даних з CSV файлу Facebook
file_path = r'D:\PHD\2023\facebook\filtered_facebook_2023.csv'
data = pd.read_csv(file_path)

# Перетворення дати в формат datetime
data['created_time'] = pd.to_datetime(data['created_time'])

# Підрахунок частоти слів 'bitcoin', 'BTC', 'біткоїн', 'биткоин', 'crypto', 'криптовалюта', 'криптокарренсі'
data['bitcoin_count'] = (
    data['text'].str.lower().str.count('bitcoin') +
    data['text'].str.lower().str.count('btc') +
    data['text'].str.lower().str.count('біткоїн') +
    data['text'].str.lower().str.count('биткоин')
)

data['crypto_count'] = (
    data['text'].str.lower().str.count('crypto') +
    data['text'].str.lower().str.count('криптовалюта') +
    data['text'].str.lower().str.count('cryptocurrency') +
    data['text'].str.lower().str.count('крипта')
)

# Групування за датою і підрахунок загальної частоти
daily_counts = data.groupby(data['created_time'].dt.date).agg({
    'bitcoin_count': 'sum', 
    'crypto_count': 'sum'
}).reset_index()
daily_counts['created_time'] = pd.to_datetime(daily_counts['created_time'])  # Перетворення на datetime

# Отримання даних про Bitcoin з yfinance
bitcoin_data = yf.download('BTC-USD', start='2023-01-01', end='2023-09-01')  # Обмеження до 1 вересня 2023 року
bitcoin_data.reset_index(inplace=True)  # Скидаємо індекс, щоб "Date" був звичайним стовпцем
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])  # Перетворення на datetime

# Об'єднання даних
merged_data = pd.merge(daily_counts, bitcoin_data[['Date', 'Close']], left_on='created_time', right_on='Date', how='outer')

# Візуалізація
plt.figure(figsize=(14, 10))

# Графік частоти для "bitcoin" (BTC)
plt.subplot(3, 1, 1)
plt.plot(merged_data['created_time'], merged_data['bitcoin_count'], label='Частота "Bitcoin", "BTC", "біткоїн", "биткоин"', color='blue')
plt.title('Частота слів "Bitcoin", "BTC", "біткоїн", "биткоин" у публікаціях Facebook')
plt.xlabel('Дата')
plt.ylabel('Кількість')
plt.xticks(rotation=45)
plt.legend()

# Графік частоти для "crypto" та "криптовалюта"
plt.subplot(3, 1, 2)
plt.plot(merged_data['created_time'], merged_data['crypto_count'], label='Частота "Crypto", "криптовалюта", "криптокарренсі"', color='orange')
plt.title('Частота слів "Crypto", "криптовалюта", "крипта", "cryptocurrency" у публікаціях Facebook')
plt.xlabel('Дата')
plt.ylabel('Кількість')
plt.xticks(rotation=45)
plt.legend()

# Комбінований графік з нормалізацією
plt.subplot(3, 1, 3)

# Нормалізація частоти
bitcoin_scaled = (merged_data['bitcoin_count'] - merged_data['bitcoin_count'].min()) / (merged_data['bitcoin_count'].max() - merged_data['bitcoin_count'].min())
crypto_scaled = (merged_data['crypto_count'] - merged_data['crypto_count'].min()) / (merged_data['crypto_count'].max() - merged_data['crypto_count'].min())
price_scaled = (merged_data['Close'] - merged_data['Close'].min()) / (merged_data['Close'].max() - merged_data['Close'].min())

# Побудова графіків
plt.plot(merged_data['created_time'], bitcoin_scaled, label='Нормалізована частота "Bitcoin"', color='blue')
plt.plot(merged_data['created_time'], crypto_scaled, label='Нормалізована частота "Crypto"', color='orange')
plt.plot(merged_data['Date'], price_scaled, label='Нормалізована ціна Bitcoin (BTC)', color='green')

plt.title('Нормалізовані значення частоти "Bitcoin", "Crypto" та ціни Bitcoin до 1 вересня 2023 року')
plt.xlabel('Дата')
plt.ylabel('Нормалізоване значення')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Завантаження даних про Bitcoin
bitcoin_data = yf.download('BTC-USD', start='2023-01-01', end='2023-09-01')  # Обмеження до 1 вересня 2023 року
bitcoin_data.reset_index(inplace=True)  # Скидаємо індекс, щоб "Date" був звичайним стовпцем
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])  # Перетворення на datetime

# Читання збережених даних
daily_counts_bitcoin = pd.read_csv(r'D:\PHD\2023\facebook\daily_counts_bitcoin.csv')
daily_counts_crypto = pd.read_csv(r'D:\PHD\2023\facebook\daily_counts_crypto.csv')

# Переконання, що обидва стовпці 'created_time' мають тип datetime
daily_counts_bitcoin['created_time'] = pd.to_datetime(daily_counts_bitcoin['created_time'])
daily_counts_crypto['created_time'] = pd.to_datetime(daily_counts_crypto['created_time'])

# Об'єднання даних
merged_data = pd.merge(daily_counts_bitcoin, daily_counts_crypto, on='created_time', how='outer', suffixes=('_bitcoin', '_crypto'))
merged_data = pd.merge(merged_data, bitcoin_data[['Date', 'Close']], left_on='created_time', right_on='Date', how='outer')

# Візуалізація
plt.figure(figsize=(14, 10))

# Графік частоти для "біткоїн"
plt.subplot(3, 1, 1)
plt.plot(merged_data['created_time'], merged_data['bitcoin_count'], label='Частота "біткоїн"', color='blue')
plt.title('Частота слів "біткоїн" у публікаціях Facebook')
plt.xlabel('Дата')
plt.ylabel('Кількість')
plt.xticks(rotation=45)
plt.legend()

# Графік частоти для "криптовалюта"
plt.subplot(3, 1, 2)
plt.plot(merged_data['created_time'], merged_data['crypto_count'], label='Частота "криптовалюта"', color='orange')
plt.title('Частота слів "криптовалюта" у публікаціях Facebook')
plt.xlabel('Дата')
plt.ylabel('Кількість')
plt.xticks(rotation=45)
plt.legend()

# Комбінований графік з нормалізацією
plt.subplot(3, 1, 3)

# Нормалізація частоти
bitcoin_scaled = (merged_data['Close'] - merged_data['Close'].min()) / (merged_data['Close'].max() - merged_data['Close'].min())
bitcoin_count_scaled = (merged_data['bitcoin_count'] - merged_data['bitcoin_count'].min()) / (merged_data['bitcoin_count'].max() - merged_data['bitcoin_count'].min())
crypto_count_scaled = (merged_data['crypto_count'] - merged_data['crypto_count'].min()) / (merged_data['crypto_count'].max() - merged_data['crypto_count'].min())

# Побудова графіків
plt.plot(merged_data['created_time'], bitcoin_count_scaled, label='Нормалізована частота "біткоїн"', color='blue')
plt.plot(merged_data['created_time'], crypto_count_scaled, label='Нормалізована частота "криптовалюта"', color='orange')
plt.plot(merged_data['Date'], bitcoin_scaled, label='Нормалізована ціна Bitcoin (BTC)', color='green')

plt.title('Нормалізовані значення частоти "біткоїн", "криптовалюта" та ціни Bitcoin до 1 вересня 2023 року')
plt.xlabel('Дата')
plt.ylabel('Нормалізоване значення')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from gensim.models import Word2Vec
import pandas as pd

# Читання даних з CSV файлу Facebook
file_path = r'D:\PHD\2023\facebook\filtered_facebook_2023.csv'
data = pd.read_csv(file_path)

# Перетворення тексту на список слів (токенізація)
sentences = [text.split() for text in data['text'].astype(str)]

# Навчання моделі Word2Vec
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python
import numpy as np

# Припустимо, що top_words_fb містить топ-10 слів Facebook
top_words_fb = ['free', 'bitcoin', 'crypto', 'business', 'people', 'bank', 'money', 'price', 'company', 'market']

# Витягти вектори слів для топ-10 слів
word_vectors = {}
for word in top_words_fb:
    if word in model.wv:
        word_vectors[word] = model.wv[word]
    else:
        print(f"Слово '{word}' не знайдено в моделі.")

# Обчислення відстаней між словами
distance_matrix = pd.DataFrame(np.zeros((len(top_words_fb), len(top_words_fb))), 
                               index=top_words_fb, 
                               columns=top_words_fb)

for i, word1 in enumerate(top_words_fb):
    for j, word2 in enumerate(top_words_fb):
        if word1 != word2:
            distance = np.linalg.norm(word_vectors[word1] - word_vectors[word2])  # Euclidean distance
            distance_matrix.iloc[i, j] = distance

# Виведення матриці відстаней
print("Відстані між словами:")
print(distance_matrix)
```

```python
# Припустимо, ваша матриця відстаней називається distance_matrix
# і вона вже визначена.

# Функція для визначення найближчих слів
def find_nearest_words(df, word, n=3):
    if word in df.index:
        distances = df[word]
        nearest_words = distances.nsmallest(n + 1).iloc[1:]  # Відкидаємо саме слово
        return nearest_words
    else:
        return None

# Визначаємо найближчі слова для кожного слова з матриці
nearest_words_dict = {}
for word in distance_matrix.index:
    nearest_words_dict[word] = find_nearest_words(distance_matrix, word)

# Виводимо результати
for word, nearest in nearest_words_dict.items():
    print(f"Найближчі слова для '{word}':")
    print(nearest)
    print()
```

```python
# Припустимо, модель зберігається в змінній model
# і у вас є список слів для пошуку найближчих слів
words_of_interest = ['free', 'bitcoin', 'crypto', 'business', 'people', 'bank', 'money', 'price', 'company', 'market']

# Словник для зберігання найближчих слів
nearest_words_dict = {}

# Функція для отримання найближчих слів
for word in words_of_interest:
    if word in model.wv:
        nearest_words = model.wv.most_similar(word, topn=5)  # Отримуємо 5 найближчих слів
        nearest_words_dict[word] = nearest_words

# Виводимо результати
for word, nearest in nearest_words_dict.items():
    print(f"Найближчі слова для '{word}':")
    for similar_word, similarity in nearest:
        print(f"  {similar_word}: {similarity:.4f}")  # Виводимо слово і його схожість
    print()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Створення графа
G = nx.Graph()

# Додавання вузлів і ребер
for word, nearest in nearest_words_dict.items():
    for similar_word, similarity in nearest:
        G.add_edge(word, similar_word, weight=similarity)

# Візуалізація графа
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # Використовуємо алгоритм для розташування вузлів
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=12, font_weight='bold', edge_color='gray')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels={k: f"{v:.2f}" for k, v in labels.items()})
plt.title('Найближчі слова в мережі')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # Додано імпорт для Seaborn

# Отримання векторів для слів "buy", "sell" та "bitcoin"
words_of_interest = ['buy', 'sell', 'bitcoin']

# Перевірка наявності слів у моделі та отримання їх векторів
word_vectors = {}
for word in words_of_interest:
    if word in model.wv:
        word_vectors[word] = model.wv[word]
    else:
        print(f"{word} не знайдено в моделі.")

# Створення DataFrame для збереження відстаней
distance_matrix = pd.DataFrame(index=words_of_interest, columns=words_of_interest)

# Обчислення відстаней між словами
for word1 in words_of_interest:
    for word2 in words_of_interest:
        if word1 in word_vectors and word2 in word_vectors:
            distance = np.linalg.norm(word_vectors[word1] - word_vectors[word2])
            distance_matrix.loc[word1, word2] = distance

# Виведення матриці відстаней
print("Відстані між словами:")
print(distance_matrix)

# Візуалізація відстаней між словами
plt.figure(figsize=(8, 6))
sns.heatmap(distance_matrix.astype(float), annot=True, cmap='Blues', square=True, cbar_kws={"shrink": .8})
plt.title('Відстані між словами "buy", "sell" та "bitcoin"')
plt.xlabel('Слова')
plt.ylabel('Слова')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Визначення слів
words = ['buy', 'sell', 'bitcoin']

# Створення графа
G = nx.Graph()

# Додавання вершин
for word in words:
    G.add_node(word)

# Додавання ребер з відстанями між словами
distances = {
    ('buy', 'sell'): model.wv.distance('buy', 'sell'),
    ('buy', 'bitcoin'): model.wv.distance('buy', 'bitcoin'),
    ('sell', 'bitcoin'): model.wv.distance('sell', 'bitcoin'),
}

for (word1, word2), distance in distances.items():
    G.add_edge(word1, word2, weight=distance)

# Візуалізація графа
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Розташування вузлів
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_color='black', font_weight='bold', edge_color='gray')

# Додавання міток до ребер з відстанями
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels={k: round(v, 2) for k, v in edge_labels.items()}, font_color='red')

plt.title('Відстані між словами "buy", "sell" та "bitcoin"')
plt.show()
```
